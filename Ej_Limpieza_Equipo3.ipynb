+ **El comando df.dropna():**

El método dropna permite, de una forma muy conveniente, filtrar los valores de una estructura de datos pandas para dejar sólo aquéllos no nulos.

```python
diab.dropna()
```

+ En este caso como la base de datos no posee datos nulos,por lo que  mantuvo la dimensión original, es decir 768 filas y 9 columnas.

Por defecto elimina las filas con 0, sin embargo si se desea eliminar las columnas con valores 0, se tiene la siguiente sintaxis:

**data_frame.dropna(axis=1)**

Si se agrega how, se puede controlar como se aplica el método, es decir: si toma el valor "all", solo se eliminarán las filas o columnas en las que todos sus elementos sean nulos. Si toma el valor "any" (valor por defecto), se eliminarán las filas o columnas en las que algún elemento sea nulo. De esta forma:



```python
diab.dropna(how="all")
diab.dropna
```

+ ### drop_duplicates()

El método drop_duplicates () de Pandas ayuda a eliminar duplicados del marco de datos.
Además se puede agregar el método keep mediante la cual se puede indicar, si se conserva la primera o ultima ocurrencia. Por defecto conserva la primera.

***Sintaxis:*** 

*DataFrame.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)*

***Nota:*** En la base de datos principal no se eliminará ningún dato duplicado, ya que eso no indica inconsistencia en el caso de estudio, sin embargo, si se tuviera una base de datos de una encuesta, con nombres repetidos, sería de gran utilidad.
